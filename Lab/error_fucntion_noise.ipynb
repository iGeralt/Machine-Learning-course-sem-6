```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random

random.seed(1)
noise = np.random.randn(10)

m = 1
x = np.linspace(1,10,10)
y = m*x + 5

plt.plot(x,y)

e = m*x + 5 + noise
plt.scatter(x,e,color = "red")
```

```python
w1 = np.linspace(-10,10,100)
w0 = np.linspace(-10,10,100)
W0, W1 = np.meshgrid (w0, w1)

l = 0
it = len(x)

for i in range(it):
    l += (W1*x[i] + W0 - e[i])**2
plt.gca().set_aspect('equal',adjustable = 'box')
plt.draw()
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('w1 vs w0 (Level graph)')

plt.contour(W0,W1,l,levels = [i for i in np.arange(-100,100,60)],colors = "red")

l1 = 0
for i in range(it):
    l1 += (W1*x[i] + W0 - y[i])**2
plt.contour(W0,W1,l1,levels = [i for i in np.arange(-200,200,60)],colors = "green")
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('Combined Surface graph')

ax.plot_surface(W0, W1, l, cmap='winter', rstride=1, cstride=1, edgecolor='none')
ax.plot_surface(W0, W1, l1, cmap='autumn', rstride=1, cstride=1, edgecolor='none')
```

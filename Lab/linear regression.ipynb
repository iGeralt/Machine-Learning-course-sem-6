```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
nx = np.linspace(0,6.2,10)
x = np.linspace(0,6.2,100)
y = np.sin(x)
e = np.sin(nx) + np.random.normal(0,0.15,10)

```

```python
#plt.plot(x,y)
plt.scatter(nx,e)
plt.plot(x,y)

```

```python
def calculate(x1,p):
    rx = []
    for i in x1:
        cx = []
        for j in range(p):
            cx.append(i**j)
        rx.append(cx)
    return pd.DataFrame(rx)
```

```python
df = calculate(nx,5)
df
```

```python
w = np.matmul(np.linalg.inv(np.matmul(np.transpose(df.values),df.values)),np.matmul(np.transpose(df.values),e))
```

```python
w
```

```python
df1 = calculate(x,5)
plt.plot(x,np.matmul(df1.values,w),color = "green")
plt.scatter(nx,e)
plt.plot(x,y,color = "red")
```

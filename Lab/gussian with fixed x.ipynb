```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from scipy.stats import norm
```

```python
nx = np.linspace(0,6.2,10)
x = np.linspace(0,6.2,100)
y = np.sin(x)
e = np.sin(nx) + np.random.normal(0,0.15,10)

```

```python
#plt.plot(x,y)
plt.scatter(nx,e)
plt.plot(x,y)

```

```python
def calculate(x1,p):
    m = np.linspace(0,6.2,p)
    #m = [1.5,5] just for 2 points
    rx = []
    for i in x1:
        cx = []
        for j in m:
            cx.append(norm.pdf(i,j,1))
        rx.append(cx)
    return pd.DataFrame(rx)
```

```python
for i in range(11):
    df = calculate(nx,i)
    w = np.matmul(np.linalg.inv(np.matmul(np.transpose(df.values),df.values)),np.matmul(np.transpose(df.values),e))

    df1 = calculate(x,i)
    plt.plot(x,np.matmul(df1.values,w),color = "green")
    #plt.show()
    plt.scatter(nx,e)
    #plt.show()
    plt.plot(x,y,color = "red")#true
    print("degree : ")
    print(i)
    plt.show()
    
```

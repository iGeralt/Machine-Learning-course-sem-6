```python
import numpy as np
import matplotlib.pyplot as plt
m = 1
x = np.linspace(1,10,11)
y = m*x
plt.plot(x,y)
plt.show()
```

```python
w1 = np.linspace(-6,9,11)
l = 0
it = len(x)
for i in range(it):
    l += (w1*x[i]- y[i])**2
```

```python
plt.plot(w1,l)
plt.xlabel('W')
plt.ylabel('Error')
plt.title('Error vs W')
plt.scatter(w1,l,c='red')
plt.show()
```

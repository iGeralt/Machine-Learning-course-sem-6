```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
m = 2
x = np.linspace(-5,5,100)
y = m*x

w0 = np.linspace(-5,5,100)

def liklihood(x,u,w0):
    l = 1
    it = len(x)
    for i in range(it):
        l *= norm.pdf(w0*x[i],u[i],1)
    return l
```

```python
plt.plot(w0,liklihood(x,y,w0))
```

```python
# with intercept
m = 2
c = 4
y1 = m*x + c

w0 = np.linspace(-10,10,100)
w1 = np.linspace(-10,10,100)

W0 , W1 = np.meshgrid(w0,w1)
```

```python
def in_liklhood(x,u,W0,W1):
    l = 1
    it = len(x)
    for i in range(it):
        l *= norm.pdf(W1*x[i]+W0,u[i],1)
    return l

def log_in_liklhood(x,u,W0,W1):
    l = 0
    it = len(x)
    for i in range(it):
        l += np.log(norm.pdf(W1*x[i]+W0,u[i],1))
    return l
```

```python
plt.gca().set_aspect('equal',adjustable = 'box')
plt.draw()
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('w1 vs w0 (Level graph)')
plt.contour(W0,W1,in_liklhood(x,y1,W0,W1))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.xlabel('w0')
plt.ylabel('w1')


ax.plot_surface(W0, W1,in_liklhood(x,y1,W0,W1) , cmap='winter', rstride=1, cstride=1, edgecolor='none')
```

```python
plt.gca().set_aspect('equal',adjustable = 'box')
plt.draw()
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('w1 vs w0 (Level graph)')
plt.contour(W0,W1,log_in_liklhood(x,y1,W0,W1))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.xlabel('w0')
plt.ylabel('w1')


ax.plot_surface(W0, W1,log_in_liklhood(x,y1,W0,W1) , cmap='winter', rstride=1, cstride=1, edgecolor='none')
```

```python
def error(W1,W0,y1):
    l = 0
    it = len(x)
    for i in range(it):
        l += (W1*x[i] + W0 - y1[i])**2
    return l
```

```python
plt.gca().set_aspect('equal',adjustable = 'box')
plt.draw()
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('w1 vs w0 (Level graph)')
plt.contour(W0,W1,error(W1,W0,y1))
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.xlabel('w0')
plt.ylabel('w1')


ax.plot_surface(W0, W1,error(W1,W0,y1) , cmap='winter', rstride=1, cstride=1, edgecolor='none')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
plt.xlabel('w0')
plt.ylabel('w1')


ax.plot_surface(W0, W1,error(W1,W0,y1) , cmap='winter', rstride=1, cstride=1, edgecolor='none')
ax.plot_surface(W0, W1,log_in_liklhood(x,y1,W0,W1) , cmap='autumn', rstride=1, cstride=1, edgecolor='none')
```
